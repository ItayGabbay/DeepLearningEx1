```python
import numpy as np
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
```

```python
with open('./costs_dropout.txt', 'r') as f:
    f.readline()
    train_costs = eval(f.readline())
    f.readline()
    valid_costs = eval(f.readline())
```

```python
train_costs = pd.DataFrame(train_costs, columns=['Iteration', 'Cost'])
```

```python
valid_costs = pd.DataFrame(valid_costs, columns=['Iteration', 'Cost'])
```

```python
%matplotlib inline
plt.plot(train_costs['Iteration'], train_costs['Cost'], label="Train")
plt.plot(valid_costs['Iteration'], valid_costs['Cost'], label="Validation")
plt.title("Costs over iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.legend()
```
